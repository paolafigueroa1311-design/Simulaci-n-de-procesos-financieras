# Simulación Montecarlo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def funcion_a_integrar(x):
    return x**2

a = 2
b = 5

N = 100

x = np.random.uniform(a, b, N)
y = np.random.uniform(0, funcion_a_integrar(b), N)
```

```python
x
```

```python
y
```

```python
n_bajo_curva=np.sum(y<funcion_a_integrar(x))
n_bajo_curva
```

```python
respuesta= (b-a)*funcion_a_integrar(b)*(n_bajo_curva/N)
respuesta
```

```python
plt.plot(x,y,'.')
plt.plot(np.linspace(a,b,1000), funcion_a_integrar(np.linspace(a,b,1000)),'r')
plt.show()
```

# Calculo de Pi

```python
# fórmula del circulo: x^2 + y^2 <=1
# simular puntos dentro de un cuadrado para ver si caen dentro del círculo
n=100000
x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)
inside = x**2 + y**2 <= 1
pi = 4 * np.sum(inside)/n
pi
```

```python
x
```

```python
y
```

```python
dentro_circulo= np.sum(inside)
dentro_circulo
```

```python
# graficar los puntos dentro del circulo
plt.figure(figsize=(6,6))
plt.scatter(x[inside], y[inside], color='blue', s=1)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f"Estimación de π: {pi}")
plt.show()
```
