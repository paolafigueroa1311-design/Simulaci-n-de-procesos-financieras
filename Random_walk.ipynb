```python
## Simulacin de una Random Walk
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n_steps =200
steps = np.random.choice([-1,1], size=n_steps)
price =100 + np.cumsum(steps)

plt.figure(figsize=(10,5))
plt.plot(price)
plt.title("Random Walk de un precio")
plt.xlabel("Tiempo")
plt.ylabel("Precio")
plt.show()
```

```python
# Distribucin de muchas simulaciones
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

n_sim =5000
n_steps =200

final_positions = []

for _ in range(n_sim):
    steps = np.random.choice([-1,1], size=n_steps)
    final_positions.append(np.sum(steps))

plt.hist(final_positions, bins=50)
plt.title("Distribucin de posiciones finales")
plt.show()
```

```python
# Incertidumbre que crece con el tiempo
import numpy as np

np.random.seed(0)

n_sim =2000
times = [10,50,100,200,400]

for t in times:
    finals = []
    for _ in range(n_sim):
        steps = np.random.choice([-1,1], size=t)
        finals.append(np.sum(steps))
print(f"t={t}, var{np.var(finals):.2f}")
```

```python
# Aproximacin a movimiento Browniano
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

T =1
n =1000
dt = T/n

increments = np.random.normal(0, np.sqrt(dt), size=n)
W = np.cumsum(increments)

plt.plot(W)
plt.title("Aproximacin a movimiento browniano")
plt.show()
```

```python
# Grafica 200 caminatas aleatorias juntas 
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)

linea=[]
n=200
tendencia= 100
for _ in range(n):
    steps = np.random.choice([-1,1], size=n)
    path = tendencia+ np.cumsum(steps)
    linea.append(path)
    plt.plot(path, alpha=0.2)

plt.title("200 Random Walks")
linea_promedio = np.mean(linea, axis=0)
plt.plot(linea_promedio, color='black', linewidth=2)
plt.xlabel("Tiempo")
plt.ylabel("Valor de caminata")
plt.legend()
plt.show
```
